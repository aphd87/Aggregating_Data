```python
transactions = pd.read_excel("grocery_database.xlsx", sheet_name= "transactions")
product_areas = pd.read_excel("grocery_database.xlsx", sheet_name= "product_areas")

transactions["sales_cost"].sum()
```

```python
transactions
```

```python
transactions = pd.merge(transactions, product_areas, how = "inner", on = "product_area_id")
```

```python
transactions
#We can see product_area_name and profit_margin are added at the end
```

```python
transactions["product_area_name"].value_counts()
```

```python
#Now you have the number of rows that the transaction table contains
#Proportion of sales are noted here
```

```python
transactions.groupby("product_area_name")["sales_cost"].sum() 
#sum of sales by product area name
#First, data frame.groupby - specify column we want to group by
#Second, in square brackets, pass variable that we want to aggregate
```

```python
transactions.groupby("product_area_name")["sales_cost"].quantile([0.25,0.5,0.75]) 
#We get 25th, 50th, 75th percentile
```

```python
#Group by returns a Pandas series rather than a data frame
#We need to create a variable around this data so that we can use it
sales_summary = transactions.groupby("product_area_name")["sales_cost"].sum() 
```

```python
sales_summary
```

```python
#If we want a data frame - we can reset the index
#Get a data frame as an output
sales_summary = transactions.groupby("product_area_name")["sales_cost"].sum().reset_index() 
```

```python
sales_summary
```

```python
#Group by multiple columns
sales_summary = transactions.groupby(["product_area_name","transaction_date"])["sales_cost"].sum().reset_index()
```

```python
sales_summary
```

```python
#We can pass a list of variables to be aggregated as well
sales_summary = transactions.groupby(["product_area_name","transaction_date"])[["sales_cost", "num_items"]].sum().reset_index()
```

```python
#Now we have multiple columns summed up
sales_summary
```

```python
#Aggregation
#Sum for sales cost column
sales_summary = transactions.groupby("product_area_name")["sales_cost"].agg("sum").reset_index()
```

```python
sales_summary
```

```python
#We can pass through more than just sum
```

```python
sales_summary = transactions.groupby("product_area_name")["sales_cost"].agg(["sum","mean"]).reset_index()
```

```python
sales_summary
```

```python
#Group by multiple columns, sum up one column by different aggregations
sales_summary = transactions.groupby(["product_area_name","transaction_date"])[["sales_cost", "num_items"]].agg(["sum", "mean"]).reset_index()
```

```python
sales_summary
```

```python
#Specify which aggregate items for which columns
sales_summary = transactions.groupby("product_area_name").agg({"sales_cost" : "sum", "num_items" : "mean"}).reset_index()
```

```python
sales_summary
```
